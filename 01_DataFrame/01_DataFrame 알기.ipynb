- row, col 동시에 선택하기
> loc, iloc 속성을 이용할 때, 콤마를 이용하여 row, col 다 명시 가능

```python
# 100, 200, 300 대상으로 cust_class, sex_type, age 가져오기 (loc 사용)
cust.loc[[100,200,300], ['cust_class', 'sex_type', 'age']]
```

```python
# 같은 형태로 iloc 사용
cust.iloc[[0,100,200], [3,4,5]] # 인덱스 번호를 적어줘야한다
```

#### 6) 조건 연산으로 데이터 선택하기 (=컬럼 조건문으로 행 추출)
- 해당 조건에 맞는 row만 선택
- 조건을 명시하고 indexing하여 가지고 옴
- ex) 남자, 3개월 평균 청구 금액이 50000 이상, 100000미만인 사람

```python
# 조건을 전부 []에 넣어주면 됨
extract = cust[(cust['sex_type'] == 'M') & (cust['r3m_avg_bill_amt'] >= 50000) & (cust['r3m_avg_bill_amt'] <= 100000)]
extract.head()
```

```python
# 조건문이 너무 길어지면 아래와 같은 방식으로 한다
# 남자
sex = cust['sex_type'] == 'M'
# 3개월 청구금액
bill = (cust['r3m_avg_bill_amt'] >= 50000) & (cust['r3m_avg_bill_amt'] < 100000)

# 합치기
cust[sex & bill].head()
```

#### 7) 정리
- 기본적으로 대괄호[]는 col을 가져오는 경우 사용하지만 slicing은 row를 가져온다
- 데이터를 가져오는 경우 loc, iloc을 사용하는데, col, row를 동시에 가져올 수 있다

### 1-2. 데이터 추가하기
#### 1) 새 column 추가하기
> - 데이터 전처리 과정에서 빈번하게 발생하는 것
> - insert 함수를 사용하여 원하는 위치에 추가

```python
# r3m_avg_bill_amt 두배로 새로운 col 생성
cust['r3m_avg_bill_amt2'] = cust['r3m_avg_bill_amt'] * 2
cust.head()
```

```python
# 기존 col을 연산하여 새로운 데이터 생성
cust['r3m_avg_bill_amt3'] = cust['r3m_avg_bill_amt2'] + cust['r3m_avg_bill_amt']
cust.head()
```

```python
# 새로운 col을 10번째 인덱스에 추가하기 (insert)
cust.insert(10, 'r3m_avg_bill_amt10', cust['r3m_avg_bill_amt'] * 10)
cust.head()
```

#### 2) column 삭제하기
> - ***drop*** 함수 사용하여 제거
> - axis는 삭제를 가로(행) 기준으로 할 것인지, 세로(열)을 기준으로 할 것인지 명시하는 파라미터임
> - 리스트를 사용하면 멀티플 col 삭제 가능

```python
# axis : dataframe은 차원이 존재함으로 항상 0과 1이 존재 (0은 행, 1은 열)
cust.drop('r3m_avg_bill_amt10', axis=1)
```

```python
# 원본을 열어보면 drop한 데이터가 존재한다
cust.head()
```

```python
# 원본 데이터를 지우고 싶다면
# 1. 데이터를 다른 데이터 프레임에 저장
cust1 = cust.drop('r3m_avg_bill_amt10', axis=1)
cust1.head()
```

```python
# 2. 원본 자체를 지우고 싶다면
# inplace 파라미터를 활용 True로 하면 원본 데이터에 수행
cust.drop('r3m_avg_bill_amt10', axis=1, inplace = True)
cust
```
