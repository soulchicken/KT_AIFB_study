# Group by
> 데이터 분할(split) -> 적용(applying) -> 데이터 병합(combine) 순으로 진행된다

## pivot, pivot_table
- pivot : pandas.pivot (index, columns, values)
- vovot table : pandas.pivot_table (index, columns, values, aggfunc), aggrefunction 함수를 사용하여 처리

- index가 2개 이상인 경우, columns가 2개 이상인 경우, 중복값이 있는 경우
- pivot_table을 사용해야한다

## stack, unstack
- stack : 컬럼 레벨에서 인덱스 레벨로 DataFrame 변경, 위에서 아래로 길게 쌓는 개념
- unstack : 인덱스 레벨에서 컬럼 레벨로 DataFrame 변경, 쌓은 것을 옆으로 늘어 놓는 개념

# Group by
### 1-1. 데이터 묶기

```python
# pandas import
import pandas as pd
import numpy as np
from IPython.display import Image
```

#### 1) 그룹화(groupby)
- 같은 값을 하나로 묶어 통계 또는 집계 결과를 얻기 위해 사용
- 아래의 세 단계로 적용하여 데이터를 그룹화, 통계에 활용
    - 데이터 분할(split) : 어떠한 기준으로 데이터를 나누기
    - operation 적용(applying) : 각 그룹에 어떤 함수를 독립적으로 적용
    - 데이터 병합(combine) : 적용되어 나온 결과들을 통합
- 데이터 분석에 있어 사용빈도가 높다
- 결과는 dictionary 형태로 나온다

```python
cust = pd.read_csv('./sc_cust_info_txn_v1.5.csv', encoding = "cp949")
cust
```

#### 1-1) groupby의 groups 속성
> - 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
gender_group = cust.groupby('sex_type')
gender_group
```

```python
gender_group.groups
```

#### 1-2) groupby 내부 함수 사용하기
- 그룹 데이터에 적용 가능한 통계 함수(NaN은 제외하여 연산)
- count : 데이터 개수
- size : 집단별 크기
- sum : 데이터의 합
- mean, std, var : 평균, 표준편차, 분산 
- min, max : 최소, 최대값

```python
# count 함수 확인
gender_group.count()
```

```python
# mean 함수 확인
gender_group.mean()
```

```python
# max값 확인
gender_group.max()
```

```python
# 특정 col 만 보는 경우 : gender별 r3m_avg_bill_amt의 평균
gender_group.mean()[['r3m_avg_bill_amt']]
```

#### 1-3) 인덱스 설정(groupby) 후 데이터 추출
- 성별 r3m_avg_bill_amt의 평균

```python
# groupby 한 상태에서 가지고 오는 경우
gender_group.mean()[['r3m_avg_bill_amt']]
```

```python
# groupby 하기 전 DataFrame에서 가지고 오는 경우
cust.groupby('sex_type').mean()[['r3m_avg_bill_amt']]
```

#### 1-4) 복수 col을 기준으로 Groupping 하기
- groupby에 column 리스트를 전달할 수 있고 복수 개의 전달도 가능
- 통계함수를 적용한 결과는 multiindex를 갖는 DataFrame

```python
# cust_class 와 sex_type 으로 index를 정하고 이에 따른 r3m_avg_bill_amt 평균 구하기
cust.groupby(['cust_class', 'sex_type']).mean()['r3m_avg_bill_amt']
```

```python
# 위와 동일하게 groupby하고 평균 구하기
multi_group = cust.groupby(['cust_class', 'sex_type'])
multi_group.mean()['r3m_avg_bill_amt']
```

```python
# index는 depth가 존재한다. (loc을 사용하여 원하는 것만 가져옴)
cust.groupby(['cust_class', 'sex_type']).mean().loc[[("D", "M")]]
```

```python
# 위 2가지를 합치면 이것처럼 만들기 가능
cust.groupby(['cust_class', 'sex_type']).mean()['r3m_avg_bill_amt'].loc[[("D", "M")]]
```

#### 1-5) index를 이용한 group by
- index가 있는 경우, groupby 함수에 level 사용 가능
- level은 index의 depth를 의미, 가장 왼쪽부터 0부터 증가
- ***set_index*** 함수
    - column 데이터를 index 레벨로 변경하는 경우
    - 기존의 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
- ***reset_index*** 함수
    - 인덱스 초기화
    - 기존의 행 인덱스 제거, 인덱스를 데이터 열로 추가

```python
# DataFrame 다시 한번 확인
cust.head()
```

#### 1-6) Multiindex를 이용한 groupping

```python
# set_index로 index 세팅 (멀티도 가능)
cust.set_index(['cust_class', 'sex_type'])
```

```python
# reset_index 활용하면 기존 DataFrame으로 변환 (set <-> reset)
cust.set_index(['cust_class', 'sex_type']).reset_index()
```

```python
# 멀티 인덱스 세팅 후 인덱스를 기준으로 groupby
# 'cust_class', 'sex_type'를 기준으로 index를 세팅하고 index를 기준으로 groupby 하는 경우
cust.set_index(['cust_class', 'sex_type']).groupby(level=[0]).mean()
```

```python
cust.set_index(['cust_class', 'sex_type']).groupby(level=[0,1]).mean()
```

#### 1-7) aggregate(집계) 함수 사용하기
groupby 결과에 집계함수를 적용하여 그룹별(mean, max 등) 데이터 확인 가능

```python
# 그룹별로 한번에 데이터를 보는 경우
cust.set_index(['cust_class', 'sex_type']).groupby(level=[0,1]).aggregate([np.mean,np.max])
```

## pivot / pivot_table 함수 활용
### 2-1. pivot
> - dataframe의 한 형태
> - 여러 분류로 섞인 ***행 데이터를 열 데이터로 회전*** 시키는 것
>     - pivot의 사전적 의미 : 회전하다
> - pivot의 형태 : pandas.pivot(index, columns, values)로 사용할 컬럼 명시

```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({'cust_id' : ['cust_1','cust_1','cust_1','cust_2','cust_2','cust_2','cust_3','cust_3','cust_3'],
                     'prod_cd' : ['p1','p2','p3','p1','p2','p3','p1','p2','p3'],
                     'grade' : ['A','A','A','A','A','A','B','B','B'],
                     'pch_amt' : [30,10,0,40,15,30,0,0,10]})
data
```

```python
# 행 : 고객 아이디, 열 : 상품 코드, 값 : 구매 금액
data.pivot(index = 'cust_id', columns = 'prod_cd', values = 'pch_amt')
```

### 2-2. Pivot_table
> - pivot_table : pandas.pivot_table(data, index, columns, aggfunc)

```python
# pivot_table을 활용하면 위와 동잏하게 만들 수 있음
data.pivot_table(index = 'cust_id', columns = 'prod_cd', values = 'pch_amt')
```

### 2-3. Pivot과 Pivot_table 차이
> - pivot이 안되는 부분을 극복할 수 있음
> - 근데 왜 둘 다 되지?

#### 1) index가 2개 이상인 경우

```python
# 피봇으로 하면 오류가 나야하는데 왜 되지?
data.pivot(index = ['cust_id', 'grade'], columns = 'prod_cd', values = 'pch_amt')
```

```python
# 피봇 테이블은 오류가 안남
data.pivot_table(index = ['cust_id', 'grade'], columns = 'prod_cd', values = 'pch_amt')
```

#### 2) columns가 2개 이상인 경우

```python
data.pivot(index = 'cust_id', columns = ['grade', 'prod_cd'], values = 'pch_amt')
```

```python
data.pivot_table(index = 'cust_id', columns = ['grade', 'prod_cd'], values = 'pch_amt')
```

#### 3) 중복 값이 있는 경우
- pivot은 중복 값이 있으면 valueError 반환
- pivot_table은 aggregation 함수를 활용하여 처리

```python
# index로 쓰인 grade가 중복
data.pivot(index='grade', columns='prod_cd', values = 'pch_amt')
```

```python
# index로 쓰인 grade가 중복 (aggfunc가 기본이 mean으로 되있다)
data.pivot_table(index='grade', columns='prod_cd', values = 'pch_amt')
```

### 2-4. pivot_table의 추가 기능 (aggregation 함수 사용)

```python
# sum으로 구하기
data.pivot_table(index='grade', columns='prod_cd', values = 'pch_amt', aggfunc=np.sum)
```

```python
# 위와 같은 결과
pd.pivot_table(data, index='grade', columns='prod_cd', values = 'pch_amt', aggfunc=np.sum)
```

```python
# mean으로 구하기
data.pivot_table(index='grade', columns='prod_cd', values = 'pch_amt', aggfunc=np.mean)
```

## stack, unstack 함수 활용

```python
df = pd.DataFrame({
    '지역' : ['서울','서울','서울','경기','경기','부산','서울','서울','부산','경기','경기','경기'],
    '요일' : ['월요일','화요일','수요일','월요일','화요일','월요일','목요일','금요일','화요일','수요일','목요일','금요일'],
    '강수량' : [100,80,1000,200,200,100,50,100,200,100,50,100],
    '강수확률' : [80,70,90,10,20,30,50,90,20,80,50,10]
})
df
```

### 3-1. stack & unstack
- stack : 컬럼 레벨에서 인덱스 레벨로 dataframe 변경
    - 데이터를 row 레벨로 쌓아올리는 개념
- unstack : 인덱스 레벨에서 컬럼 레벨로 dataframe 변경
    - stack의 반대 operation
- 둘은 역의 관계

```python
# '지역', '요일' 두 개의 인덱스를 설정
new_df = df.set_index(['지역','요일'])
new_df
```

```python
# 첫 번째 레벨의 인덱스(지역)을 컬럼으로 이동
new_df.unstack(0)
```

```python
# 두 번째 레벨의 인덱스(지역)을 컬럼으로 이동
new_df.unstack(1)
```

```python
new_df
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```
